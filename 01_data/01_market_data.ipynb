```python
import yfinance as yf
import matplotlib as plt 
import pandas as pd 

print("Fetching data...")
ticker = "SPY"
data = yf.download(ticker, start="2025-01-01", end="2026-06-25")

data['Daily_Return']= data['Adj_Close'].pct_change()

plt.figure(figsize=(10, 5))
plt.plot(data['Adj_Close'], label=f'{ticker} Adjusted Close', color = 'blue')
plt.title(f'{ticker} Price Performance')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show() 
```
